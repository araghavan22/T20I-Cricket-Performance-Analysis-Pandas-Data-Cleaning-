```python
# Question 6: Convert the above into a DataFrame with 2 columns for Country and Highest_Inns
```

```python
df_result = df.groupby('Country')['Highest_Inns_Score'].max().reset_index()
df_result.columns = ['Country', 'Highest_Inns']
df_result
```

```python
df_result.sort_values('Highest_Inns', ascending=False).head(10)
```
